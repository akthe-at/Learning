```python

import numpy as np
import pandas as pd 
import yfinance as yf
import plotly.graph_objs as plt

```

```python
tsla = yf.Ticker("TSLA")
print(tsla)
```

```python

data = yf.download(tickers="TSLA", period='1d', interval='1m')
data2 = yf.download(tickers="AAPL", period='1d', interval='1m')
data3 = yf.download(tickers="GME", period='1d', interval='1m')

the_data = [data, data2, data3]
```

```python
for data in the_data:
    fig = plt.Figure(data=[plt.Candlestick(x=data.index,
                    open=data['Open'],
                    high=data['High'],
                    low=data['Low'],
                    close=data['Close'], name = 'TESLA Live Market Data')])

    fig.show()
```
